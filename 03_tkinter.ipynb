5.1 example1 convert temperature

```python
#example convert temperature scale 
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showerror

# root window
root = tk.Tk()
root.title('Temperature Converter')
root.geometry('300x70')
root.resizable(False, False)

def fahrenheit_to_celsius(f):
    """ Convert fahrenheit to celsius
    """
    return (f - 32) * 5/9

# frame
frame = ttk.Frame(root)

# field options
options = {'padx': 5, 'pady': 5}

# temperature label
temperature_label = ttk.Label(frame, text='Fahrenheit')
temperature_label.grid(column=0, row=0, sticky='W', **options)

# temperature entry
temperature = tk.StringVar()
temperature_entry = ttk.Entry(frame, textvariable=temperature)
temperature_entry.grid(column=1, row=0, **options)
temperature_entry.focus()

# convert button
def convert_button_clicked():
    """  Handle convert button click event 
    """
    try:
        f = float(temperature.get())
        c = fahrenheit_to_celsius(f)
        result = f'{f} Fahrenheit = {c:.2f} Celsius'
        result_label.config(text=result)
    except ValueError as error:
        showerror(title='Error', message=error)

convert_button = ttk.Button(frame, text='Convert')
convert_button.grid(column=2, row=0, sticky='W', **options)
convert_button.configure(command=convert_button_clicked)

# result label
result_label = ttk.Label(frame)
result_label.grid(row=1, columnspan=3, **options)

# add padding to the frame and show it
frame.grid(padx=10, pady=10)

# start the app
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from convert_temperature import convert_to_celsius

root = tk.Tk()
root.title("Temperature Convert")
root.geometry("430x70")
root.resizable(False,False)

options = {'padx':5,'pady':5}

frame = ttk.Frame(root)
temperature_label = ttk.Label(frame,text="degrees")
temperature_label.grid(column=0,row=0,**options)

temperature = tk.StringVar()
temperature_entry = ttk.Entry(frame,textvariable=temperature)
temperature_entry.grid(column=1,row=0,**options)
temperature_entry.focus()

source_dropdown_clicked = tk.StringVar()
source_dropdown_clicked.set("fahrenhite")
drop_down_list = ["fahrenhite","kelvin"]
source_dropdown = ttk.OptionMenu(frame,source_dropdown_clicked,drop_down_list[0],*drop_down_list)
source_dropdown.grid(column=2,row=0,**options)

def convert_button_clicked():
    """Handle convert button cluck event"""
    degrees = temperature.get()
    source = source_dropdown_clicked.get()
    result = convert_to_celsius(degrees,source)
    result_label.config(text=result)
    
convert_button = ttk.Button(frame,text="Convert")
convert_button.grid(column=3,row=0,**options)
convert_button.configure(command=convert_button_clicked)

result_label = ttk.Label(frame)
result_label.grid(columnspan=4,row=1,**options)

frame.grid(padx=10,pady=10)
root.mainloop()
```

```python

# Python program to create
# a file explorer in Tkinter
  
# import all components
# from the tkinter library
from tkinter import *
from tkinter import messagebox

# import filedialog module
from tkinter import filedialog

from report_daily_tkinter import report_daily

xlsx_file_name = None
# Function for opening the
# file explorer window
def browseFiles():
    global xlsx_file_name

    filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Select a File",
                                          filetypes = (("excel files",
                                                        "*.xlsx*"),
                                                       ("all files",
                                                        "*.*")))
    xlsx_file_name = filename

    # Change label contents
    label_file_explorer.configure(text="File Opened: "+filename)
      
def making_report():
    global xlsx_file_name
    try:
      report_daily(xlsx_file_name)
      messagebox.showinfo("info", f"Making report success from file {xlsx_file_name}")    
    except Exception as e:
      print(e)
      messagebox.showerror("Error!", e)  
                                                                                                  
# Create the root window
window = Tk()
  
# Set window title
window.title('File Explorer')
  
# Set window size
window.geometry("470x150")
window.resizable(False, False)
#Set window background color

  
# Create a File Explorer label
label_file_explorer = Label(window,
                            text = "File Explorer",
                            width = 50, height = 4,
                            fg = "blue")
      
button_explore = Button(window,
                        text = "Browse Files",
                        command = browseFiles)

# field options
options = {'padx': 5, 'pady': 5}

button_report = Button(window,
                     text = "Making report", height= 2, width=50,
                     command = making_report)
  
label_file_explorer.grid(column = 1, row = 1, **options)
  
button_explore.grid(column = 2, row = 1, **options)
  
button_report.grid(column = 1,row = 2, **options)
  
# Let the window wait for any events
window.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import csv
from datetime import datetime
from collections import defaultdict

import matplotlib
matplotlib.use("TkAgg")  # ใช้ backend สำหรับ Tkinter
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# ---------------------------
# Utilities
# ---------------------------
def try_parse_date(s):
    """พยายาม parse วันที่จากหลายรูปแบบ"""
    s = s.strip()
    fmts = ["%Y-%m-%d", "%d/%m/%Y", "%m/%d/%Y", "%Y/%m/%d", "%d-%m-%Y"]
    for fmt in fmts:
        try:
            return datetime.strptime(s, fmt).date()
        except ValueError:
            continue
    # ลองให้ datetime เดา (อาจช้าหน่อยแต่ยืดหยุ่น)
    try:
        from dateutil import parser  # ถ้ามี dateutil จะยืดหยุ่นขึ้น
        return parser.parse(s, dayfirst=True).date()
    except Exception:
        raise ValueError(f"รูปแบบวันที่ไม่รู้จัก: {s}")

def read_production_csv(path):
    """
    อ่าน CSV ที่มีคอลัมน์ 'date' และ 'production'
    - ตัดช่องว่างหัวท้าย
    - ข้ามบรรทัดว่าง
    - รวม (aggregate) กรณีมีหลายแถวซ้ำวันที่เดียวกัน
    - คืนรายการ (list) ของ tuple (date, value) ที่ sort แล้ว
    """
    data = defaultdict(float)
    with open(path, "r", encoding="utf-8-sig", newline="") as f:
        reader = csv.DictReader(f)
        fieldnames_lower = [c.strip().lower() for c in reader.fieldnames or []]

        # หาแมพชื่อคอลัมน์โดยยืดหยุ่นเล็กน้อย
        def find_col(candidates):
            for c in candidates:
                if c in fieldnames_lower:
                    return fieldnames_lower.index(c)
            return None

        date_idx = find_col(["date", "วันที่", "day", "dt"])
        prod_idx = find_col(["production", "ยอดการผลิต", "output", "qty", "quantity"])

        if date_idx is None or prod_idx is None:
            raise ValueError("ไม่พบคอลัมน์ 'date' หรือ 'production' ในไฟล์ CSV")

        # สร้างชื่อคอลัมน์จริงตาม index
        cols = reader.fieldnames
        date_col = cols[date_idx]
        prod_col = cols[prod_idx]

        count_rows = 0
        for row in reader:
            count_rows += 1
            if row is None:
                continue
            date_raw = (row.get(date_col) or "").strip()
            prod_raw = (row.get(prod_col) or "").strip()
            if not date_raw:
                # ข้ามบรรทัดว่าง
                continue
            try:
                d = try_parse_date(date_raw)
            except Exception as e:
                raise ValueError(f"บรรทัดที่ {count_rows+1}: {e}")

            # แปลงยอดเป็นตัวเลข
            try:
                # รองรับคอมม่า เช่น "1,234.5"
                prod_val = float(prod_raw.replace(",", "")) if prod_raw else 0.0
            except ValueError:
                raise ValueError(f"บรรทัดที่ {count_rows+1}: production ไม่ใช่ตัวเลข -> '{prod_raw}'")

            data[d] += prod_val

    if not data:
        raise ValueError("ไฟล์ไม่มีข้อมูลที่อ่านได้")

    items = sorted(data.items(), key=lambda x: x[0])
    return items

# ---------------------------
# UI App
# ---------------------------
class ProductionApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Daily Production Chart (CSV → Graph)")
        self.geometry("900x600")

        # Top controls
        topbar = ttk.Frame(self)
        topbar.pack(side=tk.TOP, fill=tk.X, padx=8, pady=8)

        self.btn_browse = ttk.Button(topbar, text="เลือกไฟล์ CSV", command=self.on_browse)
        self.btn_browse.pack(side=tk.LEFT)

        self.chk_ma_var = tk.BooleanVar(value=False)
        self.spin_ma = tk.Spinbox(topbar, from_=2, to=30, width=5)
        ttk.Checkbutton(topbar, text="เส้นค่าเฉลี่ยเคลื่อนที่ (วัน)", variable=self.chk_ma_var, command=self.redraw).pack(side=tk.LEFT, padx=(12,4))
        self.spin_ma.pack(side=tk.LEFT)

        ttk.Button(topbar, text="รีเฟรชกราฟ", command=self.redraw).pack(side=tk.LEFT, padx=12)

        # Figure
        self.fig = Figure(figsize=(6,4), dpi=100)
        self.ax = self.fig.add_subplot(111)
        self.ax.set_title("ยังไม่มีข้อมูล - กรุณาเลือกไฟล์ CSV")
        self.ax.set_xlabel("Date")
        self.ax.set_ylabel("Production")

        self.canvas = FigureCanvasTkAgg(self.fig, master=self)
        self.canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True, padx=8, pady=8)

        # Data state
        self.series = []  # list of (date, value)

    def on_browse(self):
        path = filedialog.askopenfilename(
            title="เลือกไฟล์ CSV",
            filetypes=[("CSV files", "*.csv"), ("All files", "*.*")]
        )
        if not path:
            return
        try:
            self.series = read_production_csv(path)
            self.redraw()
        except Exception as e:
            messagebox.showerror("อ่านไฟล์ล้มเหลว", str(e))

    def redraw(self):
        self.ax.clear()
        if not self.series:
            self.ax.set_title("ยังไม่มีข้อมูล - กรุณาเลือกไฟล์ CSV")
            self.ax.set_xlabel("Date")
            self.ax.set_ylabel("Production")
            self.canvas.draw()
            return

        dates = [d for d, v in self.series]
        vals = [v for d, v in self.series]

        # เส้นหลัก
        self.ax.plot(dates, vals, marker="o", linestyle="-", label="Production")

        # ค่าเฉลี่ยเคลื่อนที่ (option)
        if self.chk_ma_var.get():
            try:
                w = int(self.spin_ma.get())
                if w >= 2 and w <= len(vals):
                    ma = []
                    s = 0.0
                    for i, x in enumerate(vals):
                        s += x
                        if i >= w:
                            s -= vals[i - w]
                        if i >= w - 1:
                            ma.append(s / w)
                        else:
                            ma.append(None)
                    ma_dates = dates
                    # plot โดยข้าม None (ทำแบบง่าย)
                    ma_x = [ma_dates[i] for i, y in enumerate(ma) if y is not None]
                    ma_y = [y for y in ma if y is not None]
                    self.ax.plot(ma_x, ma_y, linestyle="--", label=f"MA{w}")
            except Exception:
                pass

        # ตกแต่งแกน/เลย์เอาต์
        self.ax.set_title("ยอดการผลิตรายวัน")
        self.ax.set_xlabel("Date")
        self.ax.set_ylabel("Production")
        self.ax.grid(True, which="both", linestyle=":", alpha=0.5)
        self.ax.legend()

        self.fig.autofmt_xdate()  # เอียงวันที่ให้อ่านง่าย
        self.canvas.draw()

if __name__ == "__main__":
    app = ProductionApp()
    app.mainloop()
```
